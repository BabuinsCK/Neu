```python
from deeppavlov import build_model, configs

# Выбираем модель Named Entity Recognition (распознавание сущностей) для русского языка
ner_model = build_model(configs.ner.ner_rus_bert, download=True)
```

```python
text = "Меня зовут Артем, я учусь в МГТУ и живу в Москве. DeepPavlov — отличная библиотека!"
results = ner_model([text])

for tokens, tags in zip(*results):
    for token, tag in zip(tokens, tags):
        if tag != 'O':
            print(f"{token} -> {tag}")
```
